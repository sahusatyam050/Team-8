```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load all datasets
annex1 = pd.read_csv('/content/annex1.csv') # Item master
annex2 = pd.read_csv('/content/annex2.csv') # Sales
annex3 = pd.read_csv('/content/annex3.csv') # Wholesale
annex4 = pd.read_csv('/content/annex4.csv') # Loss rates

# Merge Sales with Item Master to get Categories
df = pd.merge(annex2, annex1, on='Item Code', how='left')

# Merge with Loss Rates
df = pd.merge(df, annex4[['Item Code', 'Loss Rate (%)']], on='Item Code', how='left')

# Convert Date to datetime for time-series analysis
df['Date'] = pd.to_datetime(df['Date'])

# Clean up: Handle returns (negative quantities)
df['Actual Qty'] = df.apply(lambda x: -x['Quantity Sold (kilo)'] if x['Sale or Return'] == 'return' else x['Quantity Sold (kilo)'], axis=1)
```

```python
# Aggregate volume by category
category_dist = df.groupby('Category Name')['Actual Qty'].sum().sort_values(ascending=False).reset_index()

# Plotting the distribution
plt.figure(figsize=(10, 6))
sns.barplot(data=category_dist, x='Actual Qty', y='Category Name', palette='viridis')
plt.title('Sales Volume Distribution by Category')
plt.xlabel('Total Quantity Sold (kg)')
plt.savefig('sales_distribution.png')
```

```python
# Calculate correlation for each category
correlations = []
for category in df['Category Name'].unique():
    subset = df[df['Category Name'] == category]
    corr = subset['Unit Selling Price (RMB/kg)'].corr(subset['Actual Qty'])
    correlations.append({'Category': category, 'Correlation': corr})

corr_df = pd.DataFrame(correlations).sort_values(by='Correlation')

# Plotting the correlation
plt.figure(figsize=(10, 6))
sns.barplot(data=corr_df, x='Correlation', y='Category', palette='coolwarm')
plt.title('Correlation between Price and Quantity Sold')
plt.axvline(0, color='black', linestyle='--')
plt.savefig('price_elasticity.png')
```

```python
# Merge with Wholesale Prices (Annex 3)
# Note: Matching must be done on both Date and Item Code
df_with_cost = pd.merge(df, annex3, on=['Date', 'Item Code'], how='inner')

# Calculate Net Profit per kg
# Formula: (Selling Price) - (Wholesale Price / (1 - Loss Rate))
df_with_cost['True Cost'] = df_with_cost['Wholesale Price (RMB/kg)'] / (1 - df_with_cost['Loss Rate (%)']/100)
df_with_cost['Profit per kg'] = df_with_cost['Unit Selling Price (RMB/kg)'] - df_with_cost['True Cost']

# Average profit margin by category
profit_summary = df_with_cost.groupby('Category Name')['Profit per kg'].mean().sort_values(ascending=False)
print(profit_summary)
```

```python
import pandas as pd
import numpy as np

def ingest_data():
    # Load raw data
    a1 = pd.read_csv('/content/annex1.csv')
    a2 = pd.read_csv('/content/annex2.csv')
    a3 = pd.read_csv('/content/annex3.csv')
    a4 = pd.read_csv('/content/annex4.csv')

    # Enforce standard formats across all pipelines
    a2['Date'] = pd.to_datetime(a2['Date'])
    a3['Date'] = pd.to_datetime(a3['Date'])

    # Standardize Item Codes to ensure join integrity
    for df in [a1, a2, a3, a4]:
        df['Item Code'] = df['Item Code'].astype(str).str.strip()

    return a1, a2, a3, a4

annex1, annex2, annex3, annex4 = ingest_data()
```

```python
def clean_sales_data(df_sales):
    # 1. Handle Missing Values: Drop rows with missing critical sales info
    df_sales = df_sales.dropna(subset=['Quantity Sold (kilo)', 'Unit Selling Price (RMB/kg)'])

    # 2. Handle Inconsistencies: Treat 'return' transactions as negative quantity
    df_sales['Quantity'] = np.where(df_sales['Sale or Return'] == 'return',
                                   -df_sales['Quantity Sold (kilo)'],
                                   df_sales['Quantity Sold (kilo)'])

    # 3. Handle Outliers: Cap extreme prices (Z-score method)
    # This prevents a single typo in price from ruining the category average
    price_mean = df_sales['Unit Selling Price (RMB/kg)'].mean()
    price_std = df_sales['Unit Selling Price (RMB/kg)'].std()
    df_sales = df_sales[df_sales['Unit Selling Price (RMB/kg)'] < (price_mean + 3 * price_std)]

    return df_sales

annex2_clean = clean_sales_data(annex2)
```

```python
def create_analytical_base_table(a1, a2, a3, a4):
    # Aggregating sales to a daily level per item
    daily_summary = a2.groupby(['Date', 'Item Code']).agg({
        'Quantity': 'sum',
        'Unit Selling Price (RMB/kg)': 'mean'
    }).reset_index()

    # Joining with Item Master (Annex 1)
    abt = pd.merge(daily_summary, a1, on='Item Code', how='inner')

    # Joining with Wholesale Costs (Annex 3)
    abt = pd.merge(abt, a3, on=['Date', 'Item Code'], how='inner')

    # Joining with Loss Rates (Annex 4)
    abt = pd.merge(abt, a4[['Item Code', 'Loss Rate (%)']], on='Item Code', how='left')

    # Feature Engineering: True Cost and Margin
    abt['Loss Rate (%)'] = abt['Loss Rate (%)'].fillna(0)
    abt['Effective Cost'] = abt['Wholesale Price (RMB/kg)'] / (1 - abt['Loss Rate (%)']/100)
    abt['Unit Margin'] = abt['Unit Selling Price (RMB/kg)'] - abt['Effective Cost']

    return abt

# This is your final "Clean Data" ready for all tasks in the image
master_abt = create_analytical_base_table(annex1, annex2_clean, annex3, annex4)

print("Data Engineering Complete. Prepared Structure:")
print(master_abt.head())
```

```python
import pandas as pd
from scipy import stats

# 1. Price Elasticity of Demand (Correlation)
# Grouping by item to see which specific products are most sensitive
elasticity = master_abt.groupby('Item Name').apply(
    lambda x: x['Unit Selling Price (RMB/kg)'].corr(x['Quantity'])
).sort_values()

print("Top 5 Most Price-Sensitive Items (High Elasticity):")
print(elasticity.head(5))

# 2. Statistical Significance: Weekday vs Weekend
# Does the supermarket sell significantly more on weekends?
master_abt['is_weekend'] = master_abt['Date'].dt.dayofweek.isin([5, 6])
weekend_sales = master_abt[master_abt['is_weekend']]['Quantity']
weekday_sales = master_abt[~master_abt['is_weekend']]['Quantity']

t_stat, p_val = stats.ttest_ind(weekend_sales, weekday_sales)
print(f"\nWeekend vs Weekday T-test: p-value = {p_val:.4f}")
# If p < 0.05, the difference is statistically significant.
```

```python
def detect_anomalies(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    anomalies = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return anomalies, lower_bound, upper_bound

# Detecting quantity anomalies (Panic buying or large returns)
anomalies, lb, ub = detect_anomalies(master_abt, 'Quantity')
print(f"\nDetected {len(anomalies)} anomalies in Sales Quantity.")
print(f"Typical Quantity Range: {lb:.2f} to {ub:.2f} kg")
```
